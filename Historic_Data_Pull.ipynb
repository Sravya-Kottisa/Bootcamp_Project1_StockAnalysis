```python
import pandas as pd

df = pd.read_csv('output_BTC_New.csv')
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate MAPE
mape = (abs(df['Actual'] - df['trend']) / df['Actual']).mean() * 100

# Assuming your data is in a DataFrame called 'df'
df.plot(x='time', y=['Actual', 'trend'], kind='line', 
        color=['limegreen', 'skyblue'], linewidth=2.5, figsize=(10,6))

# Add a legend
plt.legend(loc='upper right', fontsize=12)

# Customize grid
plt.grid(axis='both', linestyle='--', linewidth=0.5)

# Increase font size
plt.rcParams.update({'font.size': 12})

# Add title with subtitle
plt.title('Actual vs Predicted Values\nBTC Price (MAPE: {:.2f}%)'.format(mape), fontsize=14)

# Rotate x-axis labels
plt.xticks(rotation=45)

# Add annotations (example: maximum value)
ymax = df['Actual'].max()
#plt.annotate('Max: {:.2f}'.format(ymax), xy=(df.index[-1], ymax), 
      #       xytext=(df.index[-1], ymax+10), arrowprops=dict(facecolor='black', shrink=0.05))

# Save the graph
plt.savefig('actual_vs_predicted.png', dpi=300)

plt.xlabel('Date')
plt.ylabel('BTC Price')

# Show the plot
plt.show()
```
