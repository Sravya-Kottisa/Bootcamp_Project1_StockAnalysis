```python
# Import the required libraries and dependencies
import pandas as pd
import csv
```

```python
# Read the data from the BTC_Data_CW COPY.csv file into a Pandas DataFrame
df_BTC_Data = pd.read_csv(r"C:\Users\catpa\Documents\Group Proj 1\Starter file\BTC_Data_CW COPY.csv")
```

```python
# Calculate the avearge price of BTC in 2020
average_price = df_BTC_Data['close'].mean()

print(f"The average price of BTC in 2020 was ${average_price}")
```

```python
# Calculate the average high of BTC in 2020
import matplotlib.pyplot as plt

# Calculate the average high value of Bitcoin
average_high = df_BTC_Data['high'].mean()

# Plot the average high value
plt.figure(figsize=(16, 4))
plt.plot(df_BTC_Data['time'], df_BTC_Data['high'], label='High Value of Bitcoin')
plt.axhline(y=average_high, color='r', linestyle='--', label='Average High Value')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('High Value')
plt.title('High Value of Bitcoin and Average High Value')
plt.legend()

# Show plot
plt.show()
```

```python
# Calculate the average low value of Bitcoin
average_low = df_BTC_Data['low'].mean()

# Plot the average low value
plt.figure(figsize=(10, 6))
plt.plot(df_BTC_Data['time'], df_BTC_Data['low'], label='Low Value of Bitcoin')
plt.axhline(y=average_low, color='r', linestyle='--', label='Average Low Value')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Low Value')
plt.title('Low Value of Bitcoin and Average Low Value')
plt.legend()

# Show plot
plt.show()
```
