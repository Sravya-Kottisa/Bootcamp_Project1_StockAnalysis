```python
# Import the required libraries and dependencies
import pandas as pd
import csv
```

```python
# Read the data from the DOGE_Data.csv file into a Pandas DataFrame
df_DOGE_Data = pd.read_csv(r"C:\Users\catpa\Documents\Group Proj 1\Starter file\DOGE_Data.csv")
```

```python
# Calculate the avearge price of DOGEC in 2020
average_price = df_DOGE_Data['close'].mean()

print(f"The average price of DOGE in 2020 was ${average_price}")
```

```python
# Calculate the average high of DOGETC in 2020
import matplotlib.pyplot as plt

# Calculate the average high value of DOGE
average_high = df_DOGE_Data['high'].mean()

# Plot the average high value
plt.figure(figsize=(16, 4))
plt.plot(df_DOGE_Data['time'], df_DOGE_Data['high'], label='High Value of DOGE coin')
plt.axhline(y=average_high, color='r', linestyle='--', label='Average High Value')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('High Value')
plt.title('High Value of Bitcoin and Average High Value')
plt.legend()

# Show plot
plt.show()
```

```python
# Calculate the average low value of DOGE 
average_low = df_DOGE_Data['low'].mean()

# Plot the average low value
plt.figure(figsize=(10, 6))
plt.plot(df_DOGE_Data['time'], df_DOGE_Data['low'], label='Low Value of DOGE')
plt.axhline(y=average_low, color='r', linestyle='--', label='Average Low Value')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Low Value')
plt.title('Low Value of DOGE and Average Low Value')
plt.legend()

# Show plot
plt.show()
```
