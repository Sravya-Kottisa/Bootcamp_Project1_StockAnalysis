```python
# Calculate the avearge price of ETH in 2020
average_price = df_ETH_Data['close'].mean()

print(f"The average price of ETH in 2020 was ${average_price}")
```

```python
# Calculate the average high of ETH in 2020
import matplotlib.pyplot as plt

# Calculate the average high value of ETH
average_high = df_ETH_Data['high'].mean()

# Plot the average high value
plt.figure(figsize=(16, 4))
plt.plot(df_ETH_Data['time'], df_ETH_Data['high'], label='High Value of ETH ')
plt.axhline(y=average_high, color='r', linestyle='--', label='Average High Value')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('High Value')
plt.title('High Value of ETH and Average High Value')
plt.legend()

# Show plot
plt.show()
```

```python
# Calculate the average low value of ETH
average_low = df_ETH_Data['low'].mean()

# Plot the average low value
plt.figure(figsize=(10, 6))
plt.plot(df_ETH_Data['time'], df_ETH_Data['low'], label='Low Value of ETH')
plt.axhline(y=average_low, color='r', linestyle='--', label='Average Low Value')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Low Value')
plt.title('Low Value of ETH and Average Low Value')
plt.legend()

# Show plot
plt.show()
```
