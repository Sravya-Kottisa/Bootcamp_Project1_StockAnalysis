```python
# import Dogecoin (DOGE) csv files

doge_df = pd.read_csv('DOGE_Data.csv')

display(doge_df.head(5))
display(doge_df.tail(5))
```

```python
# Plot the "close" data for doge using "time" as the x-axis 

doge_df.set_index("time", inplace = True)

doge_df_close = doge_df["close"]

plt.figure(figsize = (16, 4))
plt.title("Dogecoin Close Prices Over Time")
plt.xlabel("Time")
plt.ylabel("Close Price")
plt.xticks(rotation=45)

doge_df_close.plot(color = "green")
```

```python
# Concat all three cryptocurrencies based on "close" and then plot into one graph.

crpyto_close_data_df = pd.DataFrame({
    'BTC_Close': btc_df['close'],
    'ETH_Close': eth_df['close'],
    'DOGE_Close': doge_df['close']
})

display(crpyto_close_data_df.head())
display(crpyto_close_data_df.tail())

crpyto_close_data_graph = crpyto_close_data_df.plot(figsize = (16, 4), 
                                                    title = "Bitcoin, Ethereum, and Dogecoin Close Prices Over Time",
                                                    style={'BTC_Close': '-', 'ETH_Close': '--', 'DOGE_Close': '-.'},
                                                    alpha=0.7)

crpyto_close_data_graph.set_xlabel("Time")
crpyto_close_data_graph.set_ylabel("Close Price (USD)")
crpyto_close_data_graph.set_yscale('log')
crpyto_close_data_graph.grid(True)

display(crpyto_close_data_graph.plot())
```

```python
# Plot the correlation between close price for btc, eth, doge

correlation_close_data_df = crpyto_close_data_df

crypto_close_correlation = correlation_close_data_df.corr()

display(crypto_close_correlation)
display(crypto_close_correlation.plot())
```

```python
# Make a new dataframe that contains "time", "close" for Bitcoin

btc_df_index_reset = btc_df.reset_index()

display(btc_df_index_reset.head())

prophet_btc_df = btc_df_index_reset[["time", "close"]]

# prophet_btc_df.set_index("time", inplace = True)

display(prophet_btc_df.head())
```

```python
# Prepare the btc data for Prophet

prophet_btc_df.columns = ["ds", "y"]

prophet_btc_df = prophet_btc_df.dropna()

prophet_btc_df.head()
```

```python
btc_prophet = Prophet()

btc_prophet
```

```python
btc_prophet.fit(prophet_btc_df)
```

```python
btc_future = btc_prophet.make_future_dataframe(periods = 72, freq = 'H')
btc_future.head()
```

```python
btc_forecast = btc_prophet.predict(btc_future)
display(btc_forecast.head())
display(btc_forecast.tail())
```

```python
btc_yhat_df = btc_forecast[["ds", "yhat"]]

plt.figure(figsize=(16, 4))
plt.plot(btc_yhat_df['ds'], btc_yhat_df['yhat'], label='Predicted yhat Price')
plt.title('Bitcoin yhat Price Prediction')
plt.xlabel('Date')
plt.ylabel('Predicted Price')
plt.grid(True)
plt.legend()
plt.show()

btc_trend_df = btc_forecast[["ds", "trend"]]

plt.figure(figsize=(16, 4))
plt.plot(btc_trend_df['ds'], btc_trend_df['trend'], label='Predicted Trend Price')
plt.title('Bitcoin Trend Price Prediction')
plt.xlabel('Date')
plt.ylabel('Predicted Price')
plt.grid(True)
plt.legend()
plt.show()
```

```python
btc_forecast.to_csv('output_BTC_forecast.csv', index=True)
```
